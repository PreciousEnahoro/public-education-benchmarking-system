```python
# Import packages and GDrive Driver for Colab to access file
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# Import data

PATH = "/content/drive/MyDrive/lea_year_model_core.csv"
df = pd.read_csv(PATH)
df.head()
```

## **Feature Engineering**

```python
# Keep only what we need
cols = [
    "LEAID","year","ca_rate","poverty_rate_5_17","median_household_income",
    "student_teacher_ratio_clean","pct_black","pct_hispanic","pct_white","pct_asian","pct_other_race",
    "urbanicity","grad_rate_2019", "State"
]
dfm = df[cols].copy()

# Make sure numeric columns are numeric
num_cols = ["ca_rate","poverty_rate_5_17","median_household_income","student_teacher_ratio_clean",
            "pct_black","pct_hispanic","pct_white","pct_asian","pct_other_race","grad_rate_2019"]
for c in num_cols:
    dfm[c] = pd.to_numeric(dfm[c], errors="coerce")

# Urban dummy
dfm["is_urban"] = (dfm["urbanicity"].astype(str).str.lower() == "urban").astype(int)

```

```python
dfm["year"] = pd.to_numeric(dfm["year"], errors="coerce")
dfm = dfm.dropna(subset=["year"])
dfm["year"] = dfm["year"].astype(int)
#dfm["ca_rate"] = StandardScaler().fit_transform(dfm[["ca_rate"]])

dfm["ca_rate"] = np.where(
    dfm["ca_rate"] > 5,   # anything above 5 is definitely wrong scale
    dfm["ca_rate"] / 100,
    dfm["ca_rate"]
)

df_pool = dfm[dfm["year"].isin([2019, 2023])].copy()

df_std = df_pool.copy()

df_std["year_2023"] = (df_std["year"] == 2023).astype(int)

#Creating interaction terms
df_std["poverty_x_2023"] = df_std["poverty_rate_5_17"] * df_std["year_2023"]
df_std["staff_x_2023"] = df_std["student_teacher_ratio_clean"] * df_std["year_2023"]

df_std.head()
```

## **Modelling**

###**Robust Model - Including State Effects**

```python
# -------------------------------
# 1. Define predictors (no is_urban)
# -------------------------------
base_cols = [
    "poverty_rate_5_17",
    "poverty_x_2023",
    "student_teacher_ratio_clean",
    "staff_x_2023",
    "pct_black",
    "pct_hispanic",
    "year_2023",
]

X = df_std[base_cols].copy()

# -------------------------------
# 2. Add state FE
# -------------------------------
state = df_std["State"].astype(str).fillna("MISSING")
state_dum = pd.get_dummies(state, drop_first=True)

X = pd.concat([X, state_dum], axis=1)

# Force numeric
X = X.apply(pd.to_numeric, errors="coerce")

# Outcome
y = pd.to_numeric(df_std["ca_rate"], errors="coerce")

# Remove inf
X = X.replace([np.inf, -np.inf], np.nan)
y = y.replace([np.inf, -np.inf], np.nan)

# Drop missing rows
mask = X.notna().all(axis=1) & y.notna()
X2 = X.loc[mask].copy()
y2 = y.loc[mask].copy()

# Add constant
X2 = sm.add_constant(X2, has_constant="add").astype(float)
y2 = y2.astype(float)

print("Rows used in final model:", len(y2))

model_state_fe = sm.OLS(y2, X2).fit(cov_type="HC3")
print(model_state_fe.summary())
```

## **Export Data For Tableau**

```python
# Predictions
pred = model_state_fe.predict(X2)

# Build frozen dataframe aligned to model rows
df_frozen = df_std.loc[mask].copy()

df_frozen["pred_ca_rate"] = pred
df_frozen["residual"] = df_frozen["ca_rate"] - df_frozen["pred_ca_rate"]
```

```python
tableau_export = df_frozen[
    [
        "LEAID",
        "State",
        "year",
        "ca_rate",
        "pred_ca_rate",
        "residual",
        "poverty_rate_5_17",
        "student_teacher_ratio_clean",
        "pct_black",
        "pct_hispanic"
    ]
].copy()

tableau_export["ca_rate"] = tableau_export["ca_rate"].round(4)
tableau_export["pred_ca_rate"] = tableau_export["pred_ca_rate"].round(4)
tableau_export["residual"] = tableau_export["residual"].round(4)

final_structural_dataset = tableau_export.copy()

final_structural_dataset.to_csv(
    "/content/drive/MyDrive/absenteeism_structural_model_FINAL.csv",
    index=False
)

print("Final dataset frozen.")
```

```python
OUT_DIR = "/content/drive/MyDrive/Absenteeism/robustmodel"  # change if you want
import os
os.makedirs(OUT_DIR, exist_ok=True)

import pickle

with open(f"{OUT_DIR}/absenteeism_robustmodel_final_HC3.pkl", "wb") as f:
    pickle.dump(model_state_fe, f)

coef_df = pd.DataFrame({
    "term": model_state_fe.params.index,
    "coef": model_state_fe.params.values,
    "se_HC3": model_state_fe.bse.values,
    "p_value": model_state_fe.pvalues.values
})
coef_df.to_csv(f"{OUT_DIR}/absenteeism_robustmodel_final_coefficients.csv", index=False)

with open(f"{OUT_DIR}/absenteeism_robustmodel_final_summary.txt", "w") as f:
    f.write(model_state_fe.summary().as_text())

#with open(f"{OUT_DIR}/absenteeism_robustmodel_final_formula.txt", "w") as f:
 #   f.write(formula_final.strip())

tableau_path = f"{OUT_DIR}/absenteeism_tableau_dataset_robustmodel.csv"
final_structural_dataset.to_csv(tableau_path, index=False)
tableau_path
```
